```python
#Corporate Bankruptcy Analysis
#Data Source: https://www.kaggle.com/datasets/fedesoriano/company-bankruptcy-prediction
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns


#Loading Data
dt = pd.read_csv("data.csv",on_bad_lines='skip')
dt.head(5)
```

```python
#Check Data Info
print("Data Info")
dt.info()
```

```python
#Check IsNull Value
print("Check IsNull Value")
dt.isnull().sum()
```

```python
#Count number of bankruptcies
print("Count number of bankruptcies")
plt.figure()
sns.countplot(x = "Bankrupt?", data = dt)
plt.show()
```

```python
#Descriptive Stastices
print("Descriptive Stastices")
dt.describe()
```

```python
#以是否破產來做資料比照
dt_is = dt.loc[dt["Bankrupt?"] == 1]
dt_non = dt.loc[dt["Bankrupt?"] == 0]
```

```python
#以Working Capital to Assets ratio 及 Debt ratio 來分析是否為宣告破產的主要原因
plt.figure(figsize=(12,5)) 
x1 = round(dt_is[' Debt ratio %'],2)
x2 = round(dt_is[' Working Capital to Total Assets'],2)

plt.scatter(x1, x2, c= 'r') 
plt.xlabel("Debt ratio %: Liability/Total Assets")
plt.ylabel("Working Capital to Total Assets")
plt.xticks(x1, rotation='vertical')
plt.title('Bankruptcy')

plt.show()

```

```python
plt.figure(figsize=(20,5)) 
x1 = round(dt_non[' Debt ratio %'],2)
x2 = round(dt_non[' Working Capital to Total Assets'],2)

plt.scatter(x1, x2, c= 'r') 
plt.xlabel("Debt ratio %: Liability/Total Assets")
plt.ylabel("Working Capital to Total Assets")
plt.xticks(x1, rotation='vertical')
plt.title('Non_Bankruptcy')

plt.show()
```

```python
#以Accounts Receivable Turnover 及 Debt ratio 來檢視有無短期變現能力(償債能力)
plt.figure(figsize=(20,5)) 
x1 = dt_is[' Accounts Receivable Turnover']
x2 = dt_is[' Debt ratio %']

plt.scatter(x1, x2, c= 'r') 
plt.xlabel("Accounts Receivable Turnover")
plt.ylabel("Debt ratio %: Liability/Total Assets")
plt.xticks(x1, rotation='vertical')
plt.title('Bankruptcy')

plt.show()
```

```python
plt.figure(figsize=(20,5)) 
x1 = dt_non[' Accounts Receivable Turnover']
x2 = dt_non[' Debt ratio %']

plt.scatter(x1, x2, c= 'r') 
plt.xlabel("Accounts Receivable Turnover")
plt.ylabel("Debt ratio %: Liability/Total Assets")
plt.xticks(x1, rotation='vertical')
plt.title('Non_Bankruptcy')

plt.show()
```
